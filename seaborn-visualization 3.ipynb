## Distribution plot of tips 

    * Distplot stands for distribution plot, it takes as input an array and plots a curve corresponding to 
      the distribution of points in the array.Distribution plots visually assess
    * The distribution of sample data by comparing the empirical distribution of the data with the theoretical values
      expected from a specified distribution



```python
sns.displot(df['tip'],bins=22,color = 'blue')
```

    Here we are know that how the tip was increased based on the total data 

```python
sns.displot(df['total_bill'],bins=22,color = 'gold')
```

```python
sns.displot(df['size'],bins=22,color = 'red')
```

```python
sns.displot(df['smoker'],color = 'purple')
```

    The graph is showing which gender has smoked highest 

```python
sns.displot(df['day'],color = 'deeppink')
```

    The graph showing us to which day has approch more to the hotel

```python
sns.displot(df['time'],bins=22,color = 'darkblue')
```

## joint plot 

    seaborn.jointplot () : Draw a plot of two variables with bivariate and univariate graphs. This function provides a convenient interface to the â€˜JointGridâ€™ class, with several canned plot kinds.
```python
sns.jointplot(x = 'total_bill',y='size',data=df,kind='reg',color = 'royalblue')
```

## KDE plot 

    KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable. It depicts the probability density at different values in a continuous variable. We can also plot a single graph for multiple samples which helps in more efficient data visualization.



```python
sns.kdeplot(df['size'],color ='blue')
```

    The size of the tabels are increased when the tip and total_bill values are increased

```python
sns.kdeplot(df['size'],color ='blue')
sns.kdeplot(df['tip'],color ='gold')
sns.kdeplot(df['total_bill'],color ='red')
```

 ## Rug plot 

A rug plot is a plot of data for a single quantitative variable, displayed as marks along an axis. It is used to visualise the distribution of the data. As such it is analogous to a histogram with zero-width bins, or a one-dimensional scatter plot.

```python
sns.rugplot(df['total_bill'],color = 'red')
```

    We are visualising the total bill value in the rug plot 

```python
sns.rugplot(df['tip'],color = 'gold')
sns.rugplot(df['total_bill'],color = 'pink')
sns.kdeplot(df['size'],color ='blue')
sns.kdeplot(df['tip'],color ='indigo')
sns.kdeplot(df['total_bill'],color ='black')
```

## Bar plot

A barplot (or barchart) is one of the most common types of graphic. It shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.

```python
sns.barplot(x='sex',y = 'tip',data=df)
```

    The highest bill was paid by male gender in the tips dataset

```python
sns.barplot(x='sex',y = 'size',data=df)
```

    The table size has increased bye the male gender

```python
sns.barplot(x='sex',y = 'total_bill',data=df)
```

    The total bill was paid by the male gender

```python
sns.barplot(x='day',y = 'size',data=df)
```

    This graph is showing about which day has recevied mostly. The day is sunday
