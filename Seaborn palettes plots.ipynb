```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df
```

```python
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day',y='total_bill',data=df,hue='sex',palette='magma')
```
