```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
```

```python
df
```

```python
fig = plt.figure(figsize = (8,5))
axes1 = fig.add_subplot(1,1,1)
axes1.hist(df['total_bill'], bins = 10, color = "blue", orientation = "vertical")
axes1.set_title("Histogram of Total bills")
axes1.set_xlabel("tip")
axes1.set_ylabel("Total Bill")
```

    Here Histogram plot about total_bill and tips how are they increasing

```python
plt.scatter(df['smoker'],df['tip'],color = 'g',marker = 'x')
plt.scatter(df['size'],df['tip'],color = 'r',marker = 'x')
```

    Here the size and smoker data are giving the conclusion about the data frame


```python
plt.plot (df['tip'],color = "blue")
plt.plot(df['sex'],color = 'red')

plt.legend()
```

     line graph about sex and tip how the two variables are increasing

```python
sns.set_style('darkgrid')
sns.pairplot(df);
```

    Here the scatter plot about the total dataframe

```python
sns.distplot(df['tip'],
            kde=False,
            hist_kws = {'alpha':1},color = 'g');
```

    The axes tick labels corresponding t the bivariate ,relational plots and NOT the distribution plots.

```python
sns.pairplot(df,diag_kws={'color':'blue','alpha':0.7});
```

```python
sns.pairplot(df,hue = 'day',palette='plasma');
```

    Data frame with palette of plasma of day column with pair plot

```python
sns.pairplot(df,hue='day',vars =['total_bill','tip'],palette = 'plasma');
```

    Here the dataframe with day combining with the total_bill and tip columns pairplot 

```python
sns.pairplot(df,
            hue='day',
            y_vars=['total_bill'],
            x_vars=['tip','size'],
            kind = 'reg',
            height=6);
```

    Here the ploting with pair about the total_bill,tip and size with the kind of reg 

```python
sns.pairplot(df,
            kind = 'reg',
            plot_kws={'ci':None,'color':'xkcd:salmon',
                     'scatter_kws':{'color':'green'}
                     }
            );
```

     Here the scatter plots with pairplot about the total dataframe
